```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
data_BM = pd.read_csv('big_mart_sales.csv')
```

```python
data_BM.head()
```

```python
# sales by outlet size
sales_by_outlet_size = data_BM.groupby('Outlet_Size').Item_Outlet_Sales.mean()

# sort by sales
sales_by_outlet_size.sort_values(inplace=True)

x = sales_by_outlet_size.index.tolist()
y = sales_by_outlet_size.values.tolist()


# set xticks
plt.xticks(labels=x, ticks=np.arange(len(x)))

plt.bar(x, y, color=['red', 'orange', 'magenta']);
```

```python
# sales by outlet size
sales_by_outlet_size = data_BM.groupby('Outlet_Size').Item_Outlet_Sales.mean()

# sort by sales
sales_by_outlet_size.sort_values(inplace=True)

x = sales_by_outlet_size.index.tolist()
y = sales_by_outlet_size.values.tolist()

# Create positions for the bars
positions = np.arange(len(x))

# set xticks
plt.xticks(ticks=positions, labels=x)

# Use the positions for the bars
plt.bar(positions, y, color=['red', 'orange', 'magenta'])

# Add axis labels and title
plt.xlabel('Outlet Size')
plt.ylabel('Sales')
plt.title('Mean Sales for each outlet type')
```

***Create a scatter plot to show the relative distribution of `Item_Weight` and `Item_Visibility`.***


---

```python
## WRITE YOUR CODE HERE
```

```python
# Create a scatter plot for Item_Weight vs Item_Visibility
plt.figure(figsize=(10, 6))

# Create the scatter plot
plt.scatter(data_BM['Item_Weight'], data_BM['Item_Visibility'],
            alpha=0.6,     # Add some transparency
            color='blue',  # Set dot color
            edgecolor='black',  # Add black edge to dots
            s=50)          # Set dot size

# Add labels and title
plt.xlabel('Item Weight', fontsize=12)
plt.ylabel('Item Visibility', fontsize=12)
plt.title('Distribution of Item Weight vs Item Visibility', fontsize=14)

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Improve layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(10, 6))

# Create the box plot using seaborn
sns.boxplot(y=data_BM['Item_Outlet_Sales'],
            color='skyblue',    # Box color
            width=0.5,          # Width of the box
            fliersize=3)        # Size of outlier points

# Add labels and title
plt.ylabel('Item Outlet Sales', fontsize=12)
plt.title('Distribution of Item Outlet Sales', fontsize=14)

# Optional: Add grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Improve layout
plt.tight_layout()

# Show the plot
plt.show()
```
