```python
### Import numpy, pandas, and get more info on the data set
## Look what the unique values are for purchase, storeID, and store. In addition, I encoded CH to be 1 and MM to be 0

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity="all"
df = pd.read_csv (r'C:\Users\rames\OneDrive\Escritorio\MMA\MMA 869 Machine Learning & Artificial Intelligence\Individual Assignment\OJ.csv')
df.info()
df.head()
df['Purchase'].unique()
df['StoreID'].unique()
df['STORE'].unique()
df['Purchase']=df['Purchase'].map({'CH':1,'MM':0})
df.info()
```

```python
pd.DataFrame(df).isna().sum() ### Check to see if there are any missing values
### I created dummies for the following features and dropped the following features accordingly 
### I dropped purchase so that my target feature did not get affected by any of my future steps taken
dummies_StoreID=pd.get_dummies(df.StoreID)
dummies_STORE=pd.get_dummies(df.STORE)
dummies_Store7=pd.get_dummies(df.Store7)
merged=pd.concat([df, dummies_StoreID, dummies_STORE], axis='columns')
merged
merged.describe().transpose()
final=merged.drop(columns=['StoreID','STORE','Purchase','Store7'])
```

```python
### The following lines of code lists the features that were created
n_features = final.shape[1]
num_col_names = list(final.select_dtypes(include=np.number).columns)
print('num_col_names: {}'.format(num_col_names))
```

```python
### I then proceeded to scale my data 

from sklearn.preprocessing import StandardScaler
final_num = final[num_col_names].to_numpy()
scaler = StandardScaler()
final_num = scaler.fit_transform(final_num)
X=final_num
```

```python
## my target feature is brought back into the code
## Please noticed that as I encoded this in my first input of this file, a 1 represents someone 
## who bought CH orange juice whereas a 0 is someone who bought a MM orange juice 
y=df['Purchase']
y.shape
y.head()
X=final
```

```python
### I imported everything I needed to be able to split the data set into train and test
### I set a random state and did a 80%/20% split for training and testing, respectively
## I imported decision tree classifier and logistic regression
## Please note that my last method will be SVM and it is imported later in this code

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.2, random_state=29)
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc, roc_auc_score, classification_report, confusion_matrix, make_scorer


############ Method 1: Decision Trees
dt= DecisionTreeClassifier(random_state=29, criterion="entropy", min_samples_split=15, max_depth=10)
dt
dt.fit(X_train,y_train)
y_pred=dt.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(roc_auc_score(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
base_dt= DecisionTreeClassifier(splitter="best", class_weight=None, random_state=29)
parameters_grid={'criterion':('gini','entropy'),
                'max_depth':[2,20,100,250,300],
                'max_leaf_nodes':[None,4,8,12,16,50],
                'min_impurity_decrease': [0,0.1,0.2],
                'min_samples_leaf':[1,10,30],
               'min_samples_split':[2,10] }
dt_gs=GridSearchCV(base_dt, param_grid=parameters_grid, scoring="roc_auc", n_jobs=2, cv=5, return_train_score=True)
%time dt_gs.fit(X_train,y_train)
```

```python
## We obtained the ROC AUC score using our best hyper parameters for our decision tree model
dt_gs.best_params_
y_pred_gs=dt_gs.predict(X_test)
print(confusion_matrix(y_test, y_pred_gs))
print(roc_auc_score(y_test,y_pred_gs))
```

```python
### The following lines will print the results of our grid search
def cv_results_to_df(cv_results):
    results=pd.DataFrame(list(cv_results['params']))
    results['mean_fit_time']=cv_results['mean_fit_time']
    results['mean_score_time']=cv_results['mean_score_time']
    results['mean_train_score']=cv_results['mean_train_score']
    results['std_train_score']=cv_results['std_train_score']
    results['mean_test_score']=cv_results['mean_test_score']
    results['std_test_score']=cv_results['std_test_score']
    results['rank_test_score']=cv_results['rank_test_score']
    results=results.sort_values(['mean_test_score'], ascending=False)
    return results
results_gs=cv_results_to_df(dt_gs.cv_results_)
results_gs
```

```python
#################### Method 2 Logistic Regression
LR=LogisticRegression()
LR.fit(X_train, y_train)
y_pred=LR.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(roc_auc_score(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
LR= LogisticRegression(class_weight=None, random_state=29)
parameters_grid={'max_iter':[10,15,50,100],
                'n_jobs':[None,4,8,12,16,50],
                'warm_start':[False,True]}
LR_gs=GridSearchCV(LR, param_grid=parameters_grid, scoring="roc_auc", n_jobs=2, cv=5, return_train_score=True)
%time LR_gs.fit(X_train,y_train)
```

```python
## After running grid search, we obtain the best hyper parameters for our Logistic Regression and 
# the AUC value we obtained using these parameters
LR_gs.best_params_
y_pred_LR=LR_gs.predict(X_test)
print(confusion_matrix(y_test, y_pred_LR))
print(roc_auc_score(y_test,y_pred_LR))
```

```python
## The following lines print the results from our grid search for our logistic regression
def cv_results_to_df(cv_results):
    results=pd.DataFrame(list(cv_results['params']))
    results['mean_fit_time']=cv_results['mean_fit_time']
    results['mean_score_time']=cv_results['mean_score_time']
    results['mean_train_score']=cv_results['mean_train_score']
    results['std_train_score']=cv_results['std_train_score']
    results['mean_test_score']=cv_results['mean_test_score']
    results['std_test_score']=cv_results['std_test_score']
    results['rank_test_score']=cv_results['rank_test_score']
    results=results.sort_values(['mean_test_score'], ascending=False)
    return results
results_LR=cv_results_to_df(LR_gs.cv_results_)
results_LR
```

```python
####################### Method 3 SVM
#SVM
from sklearn import svm
svm_model=svm.SVC(kernel='linear')
svm_model.fit(X_train, y_train)
```

```python

y_pred=svm_model.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(roc_auc_score(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
svm_model= svm.SVC(class_weight=None, random_state=29)
parameters_grid={'C': [1, 10, 100, 1000], 'gamma': [0.001, 0.0001], 'kernel': ['linear','rbf']}
 
svm_model_gs=GridSearchCV(svm_model, param_grid=parameters_grid, scoring="roc_auc", n_jobs=2, cv=5, return_train_score=True)
%time svm_model_gs.fit(X_train,y_train)
```
