#### `Consider the following dataframe`

---

```python
sample_df = pd.DataFrame({
    'id'   :  [1, 2, 3, 4, 5],
    'name' :  [ 'Akash', 'Akansha', 'Naman', 'Pulkit', 'Anand'],
    'marks':  [ 22, 23, 43, 41, 12]
})
sample_df
```

#### Sort the  dataframe by `marks` column in descending order

---

```python
# WRITE YOUR CODE HERE
```

```python
# Sort by marks in descending order
sorted_df = sample_df.sort_values(by='marks', ascending=False)

# Display the sorted DataFrame
sorted_df
```

#### Consider another dataframe city
---

```python
city = pd.DataFrame({
    'name' : ['Akash', 'Akansha', 'Pulkit', 'Anand', 'Naman'],
    'city' : ['Delhi', 'Gurugram', 'Gurugram', 'Delhi', 'Noida']
})
city
```

***Use the dataframe `sample_df` and `city` and find out the marks of students who live in Gurugram.***

---

```python
### Write your code here
```

```python
# Merge the two DataFrames on the 'name' column
merged_df = pd.merge(sample_df, city, on='name')

# Filter for students who live in Gurugram
gurugram_students = merged_df[merged_df['city'] == 'Gurugram']

# Select the relevant columns
result = gurugram_students[['name', 'marks', 'city']]

# Display the result
result
```

#### `READ THE BIG MART SALES DATA`

---

```python
big_mart_data = pd.read_csv('big_mart_sales.csv')
```

```python
big_mart_data.head()
```

#### Find out the number of unique `Item_Type` in each of the `Outlet_Identifier`

---

```python
# WRITE YOUR CODE HERE
```

```python
# Count unique Item_Types for each Outlet_Identifier
outlet_item_types = big_mart_data.groupby('Outlet_Identifier')['Item_Type'].nunique()

# Display the results
outlet_item_types
```

#### Find out the average `Item_Outlet_Sales` for each category of `Outlet_Size`

---

```python
# WRITE YOUR CODE HERE
```

```python
# Calculate average Item_Outlet_Sales for each Outlet_Size category
avg_sales_by_size = big_mart_data.groupby('Outlet_Size')['Item_Outlet_Sales'].mean()

# Display the results
avg_sales_by_size
```

#### Select only the categorical variables from the dataframe and use `get_dummies` function to convert those into the numerical ones.

---

```python
# WRITE YOUR CODE HERE
```

```python
# Step 1: Select only categorical columns
# Assuming categorical columns are those with 'object' or 'category' dtype
categorical_cols = big_mart_data.select_dtypes(include=['object', 'category']).columns

# Step 2: Create a dataframe with only categorical columns
categorical_data = big_mart_data[categorical_cols]

# Step 3: Convert categorical variables to dummy/indicator variables
dummy_data = pd.get_dummies(categorical_data, drop_first=True)
```

```python
dummy_data
```

```python
# Step 4: If you want to join these back to the original numerical columns
numerical_cols = big_mart_data.select_dtypes(exclude=['object', 'category']).columns
numerical_data = big_mart_data[numerical_cols]

# Step 5: Combine numerical and one-hot encoded data
final_data = pd.concat([numerical_data, dummy_data], axis=1)

# Display the first few rows of the transformed data
final_data.head()
```
